```python
df1 = pd.read_csv('toddlers/Toddler Autism dataset July 2018.csv',na_values='?')
```

```python
df1.drop(['Case_No','Who completed the test'], axis=1, inplace=True)
df1.columns
```

```python
df1.dtypes
```

```python
df1.drop('Qchat-10-Score', axis = 1, inplace = True)
```

```python
df1.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
columns = ['Ethnicity', 'Family_mem_with_ASD', 'Class/ASD Traits ', 'Sex', 'Jaundice']
for col in columns:
    df1[col] = le.fit_transform(df1[col])
df1.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1.drop(['Class/ASD Traits '], axis = 1)
Y = df1['Class/ASD Traits ']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state = 7)
```

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import accuracy_score
```

```python
model=SVC()
model.fit(x_train, y_train)
pred = model.predict(x_test).astype(int)
print(accuracy_score(y_test,pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svc = SVC()

params = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}

clf = GridSearchCV(svc, param_grid = params, scoring = 'accuracy', cv = 10, verbose = 2)

clf.fit(x_train, y_train)
clf.best_params_
```

```python
svc1 = SVC(C = 0.8, gamma = 0.1, kernel = 'linear')
svc1.fit(x_train, y_train)
pred = svc1.predict(x_test)
print(accuracy_score(y_test, pred))
```
